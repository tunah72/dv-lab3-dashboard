# Tiền xử lý dữ liệu (Preprocessing)
Khắc phục lỗi tìm thấy ở bước EDA và lưu vào `data/cleaned/`.

```python
import pandas as pd
import os
import glob

raw_path = '../data/raw/'
clean_path = '../data/cleaned/'
os.makedirs(clean_path, exist_ok=True)
csv_files = glob.glob(os.path.join(raw_path, '*.csv'))
dfs = {os.path.basename(f).replace('.csv', ''): pd.read_csv(f) for f in csv_files}
```

## 1. Sửa lỗi Dữ liệu bất thường (Anomalies)

```python
if 'order_items' in dfs:
    # Xóa các dòng có giá trị âm hoặc 0
    df_oi = dfs['order_items']
    dfs['order_items'] = df_oi[(df_oi['qty'] > 0) & (df_oi['price'] > 0)].copy()
    print('Đã làm sạch bảng order_items.')
```

## 2. Đồng bộ kiểu dữ liệu (Date & String IDs)

```python
for name, df in dfs.items():
    for col in df.columns:
        if 'date' in col.lower():
            df[col] = pd.to_datetime(df[col], errors='coerce')
        elif col.endswith('_id'):
            df[col] = df[col].astype(str)
    print(f'Đã đồng bộ kiểu dữ liệu cho bảng {name}.')
```

## 3. Lưu dữ liệu đã làm sạch

```python
for name, df in dfs.items():
    save_path = os.path.join(clean_path, f'{name}.csv')
    df.to_csv(save_path, index=False)
    print(f'Đã lưu: {save_path}')
```
