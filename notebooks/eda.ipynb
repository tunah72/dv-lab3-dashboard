# Khám phá Dữ liệu (EDA)
Notebook này đọc 12 files từ `data/raw/` để phát hiện lỗi.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob

raw_path = '../data/raw/'
csv_files = glob.glob(os.path.join(raw_path, '*.csv'))
print(f'Tìm thấy {len(csv_files)} files CSV.')
```

```python
dfs = {}
for f in csv_files:
    name = os.path.basename(f).replace('.csv', '')
    dfs[name] = pd.read_csv(f)
    print(f'Bảng {name.upper()}: {dfs[name].shape[0]} dòng, {dfs[name].shape[1]} cột')
```

## 1. Phát hiện Missing Values

```python
for name, df in dfs.items():
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        print(f'\n--- Bảng {name.upper()} ---')
        print(missing)
```

## 2. Tìm kiếm Outliers và Anomalies

```python
if 'order_items' in dfs:
    display(dfs['order_items'].describe())
    invalid_qty = dfs['order_items'][dfs['order_items']['qty'] <= 0]
    print(f'Phát hiện {len(invalid_qty)} dòng có số lượng <= 0.')
    invalid_price = dfs['order_items'][dfs['order_items']['price'] <= 0]
    print(f'Phát hiện {len(invalid_price)} dòng có đơn giá <= 0.')
```
